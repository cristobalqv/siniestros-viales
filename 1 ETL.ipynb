```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Función para contar los nulos

def verificar_tipo_datos(df):
    diccionario = {'nombre_campo':[], 'tipo_dato':[], '%_no_nulos':[], '%_nulos':[], 'nulos':[]}
    for columna in df.columns:
        porcentaje_no_nulos = (df[columna].count() / len(df[columna])) *100
        diccionario['nombre_campo'].append(columna)
        diccionario['tipo_dato'].append(df[columna].apply(type).unique())
        diccionario['%_no_nulos'].append(round(porcentaje_no_nulos, 2))
        diccionario['%_nulos'].append(round(100-porcentaje_no_nulos, 2))
        diccionario['nulos'].append(df[columna].isna().sum())
    df_info = pd.DataFrame(diccionario)
    return df_info
```

```python
# para añadir un rango de edad para nuestro analisis creamos una funcion que define un rango de edad
def rango_edad(edad):
    if edad == 'SD':
        return np.nan
    elif edad < 18:
        return "0 a 17 años"
    elif 18 <= edad <= 29:
        return "18 a 29 años"
    elif 30 <= edad <= 39:
        return "30 a 39 años"
    elif 40 <= edad <= 49:
        return "45 a 49 años"
    elif 50 <= edad <= 59:
        return "50 a 59 años"
    elif edad >= 60:
        return "60 años o mas"
    
```

```python
# aplicamos esta funcion a la columna 'EDAD'
dfvictimas['RANGO EDAD'] = dfvictimas['EDAD'].apply(rango_edad)
```

```python
# cambiamos nombre de columna
dfvictimas = dfvictimas.rename(columns={'ID_hecho':'ID_HECHO'})
```
