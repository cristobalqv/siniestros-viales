```python
# Librer√≠as utilizadas en el documento
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# remueve los warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
dfhomicidios = pd.read_csv('C:\\Users\\cquir\\OneDrive\\Escritorio\\Data Science SH\\Proyecto Individual 2\\bases de datos\\homicidios.csv')
dfvictimas = pd.read_csv('C:\\Users\\cquir\\OneDrive\\Escritorio\\Data Science SH\\Proyecto Individual 2\\bases de datos\\victimas.csv')
```

```python
# convierto las columnas a datos categoricos
dfhomicidios['TIPO_DE_CALLE'] = pd.Categorical(dfhomicidios['TIPO_DE_CALLE'], categories=dfhomicidios['TIPO_DE_CALLE'].unique(), ordered=True)
dfhomicidios['HORA'] = pd.Categorical(dfhomicidios['HORA'], categories=dfhomicidios['HORA'].unique(), ordered=True)
dfhomicidios['VICTIMA'] = pd.Categorical(dfhomicidios['VICTIMA'], categories=dfhomicidios['VICTIMA'].unique(), ordered=True)
dfhomicidios['ACUSADO'] = pd.Categorical(dfhomicidios['ACUSADO'], categories=dfhomicidios['ACUSADO'].unique(), ordered=True)
```

```python
# Crear una figura con dos subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 6))

# Gr√°fico de barras para la variable tipo_de_calle
dfhomicidios['TIPO_DE_CALLE'].value_counts().plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('TIPO_DE_CALLE')

# Gr√°fico de barras para los ingresos
dfhomicidios['HORA'].value_counts().plot(kind='bar', ax=axes[0,1], color='salmon')
axes[0,1].set_title('HORA')

# Gr√°fico de barras para los ingresos
dfhomicidios['VICTIMA'].value_counts().plot(kind='bar', ax=axes[1,0], color='green')
axes[1,0].set_title('VICTIMA')

# Gr√°fico de barras para los ingresos
dfhomicidios['ACUSADO'].value_counts().plot(kind='bar', ax=axes[1,1], color='yellow')
axes[1,1].set_title('ACUSADO')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar los gr√°ficos
plt.show()
```

```python
# A√ëOS, MESES Y DIAS

# Crear el gr√°fico
plt.figure(figsize=(10, 7))

# Subplot 1: Count plot para los meses de la columna FECHA
plt.subplot(2, 2, 1)  
sns.countplot(x=dfhomicidios['FECHA'].dt.year, palette='deep')
plt.title('Homicidios por a√±o')
plt.xlabel('A√±o')
plt.ylabel('Cantidad')

# Subplot 2: Count plot para los d√≠as de la semana de la columna FECHA
plt.subplot(2, 2, 2)  
sns.countplot(x=dfhomicidios['FECHA'].dt.month , palette='deep')
plt.title('Homicidios por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad')

# Subplot 3: Count plot para las horas de la columna HORA
plt.subplot(2, 2, 3)  
sns.countplot(x=dfhomicidios['FECHA'].dt.dayofweek, palette='deep')
plt.title('Homicidios por d√≠a de la semana')
plt.xlabel('D√≠a de la semana')
plt.ylabel('Cantidad')

# Subplot 4: Count plot para los a√±os de la columna FECHA
plt.subplot(2, 2, 4)  
sns.countplot(x=dfhomicidios['HORA'], palette='deep')
plt.title('Homicidios por hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
